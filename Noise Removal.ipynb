```python
#Noise Removal By Morphology Operations

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("2.jpg")

# Check if the image was loaded successfully
if image is None:
    print("Error: Image could not be loaded.")
else:
    # Rescale the image (resize) to a specific width and height
    scale_percent = 50  # Percent of original size
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)

    # Resize the image
    resized_image = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)

    # Convert to grayscale
    gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

    # Apply Otsu's binarization
    _, otsu_thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Morphological Operations for Noise Removal
    # Define a kernel (structuring element) for morphological operations
    kernel = np.ones((3, 3), np.uint8)  # 3x3 kernel

    # Apply morphological opening (erosion followed by dilation) to remove noise
    opened_image = cv2.morphologyEx(otsu_thresh, cv2.MORPH_OPEN, kernel)

    # Apply morphological closing (dilation followed by erosion) to fill small holes
    closed_image = cv2.morphologyEx(otsu_thresh, cv2.MORPH_CLOSE, kernel)

    # Display the results
    plt.rcParams['figure.figsize'] = (16, 8)  # Adjust figure size
    plt.subplot(141), plt.imshow(gray, cmap='gray'), plt.title('Resized Grayscale Image')
    plt.subplot(142), plt.imshow(otsu_thresh, cmap='gray'), plt.title("Otsu's Binarization")
    plt.subplot(143), plt.imshow(opened_image, cmap='gray'), plt.title("Opened (Noise Removed)")
    plt.subplot(144), plt.imshow(closed_image, cmap='gray'), plt.title("Closed (Holes Filled)")
    plt.show()

    # Print the optimal threshold value using Otsu's method
    print(f"Otsu's optimal threshold value: {_}")
```

```python
#Noise Removal By Small Dots

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("2.jpg")

# Check if the image was loaded successfully
if image is None:
    print("Error: Image could not be loaded.")
else:
    # Rescale the image (resize) to a specific width and height
    scale_percent = 50  # Percent of original size
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)

    # Resize the image
    resized_image = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)

    # Convert to grayscale
    gray = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

    # Apply Otsu's binarization
    _, otsu_thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Morphological Opening: Remove small dots (noise)
    kernel = np.ones((5, 5), np.uint8)  # Structuring element (5x5 kernel)
    opened_image = cv2.morphologyEx(otsu_thresh, cv2.MORPH_OPEN, kernel)

    # Find contours to remove small connected components (small dots)
    contours, _ = cv2.findContours(opened_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Create a blank image to draw only the large components
    cleaned_image = np.zeros_like(opened_image)
    
    # Filter contours based on area (remove small dots)
    min_area = 100  # Minimum area threshold to keep objects
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area >= min_area:  # Keep contours with area larger than the threshold
            cv2.drawContours(cleaned_image, [cnt], -1, (255), thickness=cv2.FILLED)
    
    # Display the results
    plt.rcParams['figure.figsize'] = (16, 8)  # Adjust figure size
    plt.subplot(141), plt.imshow(gray, cmap='gray'), plt.title('Resized Grayscale Image')
    plt.subplot(142), plt.imshow(otsu_thresh, cmap='gray'), plt.title("Otsu's Binarization")
    plt.subplot(143), plt.imshow(opened_image, cmap='gray'), plt.title("Opened (Noise Removed)")
    plt.subplot(144), plt.imshow(cleaned_image, cmap='gray'), plt.title("Dots Removed")
    plt.show()

    # Print the optimal threshold value using Otsu's method
    print(f"Otsu's optimal threshold value: {_}")
```

```python
#Noise Removal By Deskewing

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to compute the skew angle of an image
def get_skew_angle(image):
    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Threshold the image to binary (use adaptive thresholding or simple binary threshold)
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    # Find the edges of the binary image
    edges = cv2.Canny(thresh, 50, 150, apertureSize=3)

    # Use Hough Line Transform to detect lines in the image
    lines = cv2.HoughLines(edges, 1, np.pi / 180, 200)
    
    # Calculate the skew angle by averaging the angles of detected lines
    if lines is not None:
        angles = []
        for line in lines:
            rho, theta = line[0]
            angle = np.degrees(theta) - 90  # Convert from radians to degrees
            angles.append(angle)
        
        # Return the median angle to avoid extreme values
        return np.median(angles)
    else:
        return 0  # No skew

# Function to deskew the image
def deskew_image(image):
    # Get the skew angle
    angle = get_skew_angle(image)
    print(f"Skew angle: {angle} degrees")
    
    # Get the rotation matrix to correct the skew
    height, width = image.shape[:2]
    center = (width // 2, height // 2)
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    
    # Perform the rotation (deskewing)
    deskewed_image = cv2.warpAffine(image, rotation_matrix, (width, height), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)
    
    return deskewed_image

# Load the image
image = cv2.imread("2.jpg")

# Check if the image was loaded successfully
if image is None:
    print("Error: Image could not be loaded.")
else:
    # Deskew the image
    deskewed_image = deskew_image(image)

    # Display the original and deskewed images
    plt.rcParams['figure.figsize'] = (16, 8)  # Adjust figure size
    plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
    plt.subplot(122), plt.imshow(cv2.cvtColor(deskewed_image, cv2.COLOR_BGR2RGB)), plt.title('Deskewed Image')
    plt.show()
```

```python
#Noise Removal By Orientation/Rotation

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to detect the orientation (rotation angle) of the image
def detect_orientation(image):
    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Threshold the image (binary image for edge detection)
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    # Detect edges using Canny edge detector
    edges = cv2.Canny(thresh, 50, 150, apertureSize=3)

    # Use Hough Line Transform to detect lines in the image
    lines = cv2.HoughLines(edges, 1, np.pi / 180, 100)
    
    # Compute the angle of the lines detected
    if lines is not None:
        angles = []
        for line in lines:
            rho, theta = line[0]
            angle = np.degrees(theta) - 90  # Convert from radians to degrees
            angles.append(angle)

        # Return the median angle to avoid extreme values
        return np.median(angles)
    else:
        return 0  # No lines detected

# Function to rotate the image based on detected orientation
def rotate_image(image, angle):
    # Get the center of the image
    center = (image.shape[1] // 2, image.shape[0] // 2)

    # Get the rotation matrix
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    
    # Rotate the image
    rotated_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]),
                                   flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)
    
    return rotated_image

# Load the image
image = cv2.imread("2.jpg")

# Check if the image was loaded successfully
if image is None:
    print("Error: Image could not be loaded.")
else:
    # Detect the orientation angle of the image
    angle = detect_orientation(image)
    print(f"Detected rotation angle: {angle} degrees")
    
    # Rotate the image based on the detected angle
    rotated_image = rotate_image(image, angle)
    
    # Display the original and rotated (oriented) images
    plt.rcParams['figure.figsize'] = (16, 8)  # Adjust figure size
    plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
    plt.subplot(122), plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)), plt.title('Oriented Image')
    plt.show()
```

```python
# Noise removal by Perspective Transformation

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform perspective transformation
def perspective_transform(image):
    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Threshold the image to get a binary image (can use adaptive or simple thresholding)
    _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
    
    # Find the contours in the thresholded image
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Find the largest contour (which should correspond to the document or object of interest)
    largest_contour = max(contours, key=cv2.contourArea)
    
    # Approximate the contour to a polygon (we expect a quadrilateral)
    epsilon = 0.02 * cv2.arcLength(largest_contour, True)
    approx = cv2.approxPolyDP(largest_contour, epsilon, True)
    
    # Ensure we have four points (the corners of the object/document)
    if len(approx) == 4:
        pts1 = np.float32(approx)  # The four points from the image
        # Define the points for the desired top-down view (order: top-left, top-right, bottom-right, bottom-left)
        width = int(max(np.linalg.norm(pts1[0] - pts1[1]), np.linalg.norm(pts1[2] - pts1[3])))
        height = int(max(np.linalg.norm(pts1[0] - pts1[3]), np.linalg.norm(pts1[1] - pts1[2])))
        pts2 = np.float32([[0, 0], [width - 1, 0], [width - 1, height - 1], [0, height - 1]])

        # Compute the perspective transformation matrix
        matrix = cv2.getPerspectiveTransform(pts1, pts2)
        
        # Apply the perspective transformation
        transformed_image = cv2.warpPerspective(image, matrix, (width, height))
        
        return transformed_image
    else:
        print("No quadrilateral contour detected.")
        return image  # If no quadrilateral is found, return the original image

# Load the image
image = cv2.imread("2.jpg")

# Check if the image was loaded successfully
if image is None:
    print("Error: Image could not be loaded.")
else:
    # Perform perspective transformation
    transformed_image = perspective_transform(image)
    
    # Display the original and transformed images
    plt.rcParams['figure.figsize'] = (16, 8)  # Adjust figure size
    plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
    plt.subplot(122), plt.imshow(cv2.cvtColor(transformed_image, cv2.COLOR_BGR2RGB)), plt.title('Transformed Image')
    plt.show()
```

```python
# Noise Removal with Border Removal

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to remove the border
def remove_border(image):
    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Threshold the image to get a binary image (white object on black background)
    _, thresh = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY)
    
    # Invert the binary image (black object on white background)
    thresh = cv2.bitwise_not(thresh)
    
    # Find contours in the thresholded image
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Get the bounding box of the largest contour (which should be the main content)
    x, y, w, h = cv2.boundingRect(contours[0])
    
    # Crop the image to remove the border based on the bounding box
    cropped_image = image[y:y+h, x:x+w]
    
    return cropped_image

# Load the image
image = cv2.imread("2.jpg")

# Check if the image was loaded successfully
if image is None:
    print("Error: Image could not be loaded.")
else:
    # Remove the border
    image_no_border = remove_border(image)
    
    # Display the original and border-removed images
    plt.rcParams['figure.figsize'] = (16, 8)  # Adjust figure size
    plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
    plt.subplot(122), plt.imshow(cv2.cvtColor(image_no_border, cv2.COLOR_BGR2RGB)), plt.title('Image Without Border')
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to remove noise and add padding
def remove_noise_and_add_padding(image, padding=50):
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Threshold the image to get a binary image (white object on black background)
    _, thresh = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY_INV)
    
    # Invert the binary image (black object on white background)
    thresh = cv2.bitwise_not(thresh)
    
    # Apply morphological operations to remove small noise (dots)
    kernel = np.ones((5, 5), np.uint8)
    thresh = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)  # Close operation (dilation followed by erosion)
    
    # Find contours to detect the boundaries of the object
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Check if contours were found
    if contours:
        # Get the bounding box of the largest contour (which should be the main content)
        x, y, w, h = cv2.boundingRect(contours[0])
        
        # Crop the image to remove the noise (based on the bounding box)
        cropped_image = image[y:y+h, x:x+w]
        
        # Add padding around the cropped image
        padded_image = cv2.copyMakeBorder(cropped_image, padding, padding, padding, padding, cv2.BORDER_CONSTANT, value=(255, 255, 255))
        
        return padded_image
    else:
        print("No contours found in the image.")
        return image  # Return the original image if no contours are found

# Load the image
image = cv2.imread("2.jpg")

# Check if the image was loaded successfully
if image is None:
    print("Error: Image could not be loaded.")
else:
    # Remove noise and add padding
    image_no_noise_with_padding = remove_noise_and_add_padding(image)
    
    # Display the original and processed images
    plt.rcParams['figure.figsize'] = (16, 8)  # Adjust figure size
    plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
    plt.subplot(122), plt.imshow(cv2.cvtColor(image_no_noise_with_padding, cv2.COLOR_BGR2RGB)), plt.title('Image With Noise Removed and Padding')
    plt.show()
```
